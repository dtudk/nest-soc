# Installation

In this tutorial, we explain how one can install the nest-soc package.

## Manual installation

To install the nest-soc, first download the newest version from the official github page (https://github.com/dtudk/nest-soc)

### To download:
1. Click on button "<> Code"
2. Choose download ZIP

### To install:
1. Unzip the file
2. Open the unzipped folder using VS Code (https://code.visualstudio.com/download).
3. With the terminal in the root folder (i.e., the one containing the file pyproject.toml) type the command:

        python3 -m build

> The command should create a folder called dist, containing a file named "nest_soc-X.Y.Z.tar.gz" where X, Y and Z are the version numbers

4. In the same terminal, type the following command to install the package and dependencies **substituting the X, Y, and Z** the version you downloaded:

        python3 install -/dist/nest_soc-X.Y.Z.tar.gz

Done, python should install the nest-soc and now you can access it in python as **nest**
